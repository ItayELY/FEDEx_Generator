```python
import Measures.Bins as Bins
from utils import is_numeric

class EvenBin(Bins.UserBin):
    def __init__(self, source_column, result_column):
        super().__init__(source_column, result_column)
        if source_column is None:
            self.binned_source = None
        else:
            self.binned_source = source_column.apply(lambda x: x % 2)

        self.binned_result = result_column.apply(lambda x: x % 2)


    def get_binned_source_column(self):
        return self.binned_source

    def get_binned_result_column(self):
        return self.binned_result

    def get_bin_name(self):
        return self.result_column.name

    def get_bin_representation(self, item):
        return "Odd" if item == 1 else "Even"


def bin_columns(source_col, result_col):
    if result_col.dtype.name not in ['int64', 'int32', 'int']:
        return []

    return [EvenBin(source_col, result_col)]


Bins.Bins.register_binning_method(bin_columns, use_only_user_bins=True)
```
